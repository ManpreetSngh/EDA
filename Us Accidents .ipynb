# Findings


1. Los Angeles was the city with the highest number of accidents but take note that there was no data from New York (the most      populated city in the US) .
2. Less than 5% of cities have more than 1000 yearly accidents
3. Over 1100 cities have reported only one accident
4. After midnight, little high  no. of accidents occur during 6-8 am (probably the reason behind is people leave for work during    that time period) but the next highest peak of accident is during 3-5 pm (probably the reason behind is people leave for        home after work during that time period)
5. No. of accidents during weekends are less as compared to week days but During Weekends the peak of accidents is during 
   12 noon - 6 pm unlike weekdays .
6. 4.6 % of accidents found to be very Severe
7. Data for 2017 , 2018 seems more balanced, while there were missing data in 2016 and 2020 (2016 start of data collection so      system might not have been fully in place).
8. Accidents occurence tend to drop during summer months & pick up by Winter
9. Accidents Happend Most frequently on right side of road
10. Surprisingly, most accidents happened when the weather was 'fair'. Perhaps weather (bad weather) was not a big contributing      factor to accidents.
