```python
### Check Stats of a team 
def team(x):
    print(x)
    b=ipl[ipl['Team1'] == x].shape[0]
    c= ipl[ipl['Team2'] == x].shape[0]
    print('No. of MAtches Played : ',b+c)
    d=ipl[ipl['WinningTeam'] == x].shape[0]
    print('No.of Matches Won :',d)
    Won_percentage=(d/(b+c))*100
    print('Winning Percentage :',np.round(Won_percentage,2))
    return  
team('Punjab Kings')
```

```python
### created a function to compare two teams stats
def compare(a,b):
    h=(a,b)
    e=ipl[(ipl.Team1.isin(h))&(ipl.Team2.isin(h))]
    l=e['WinningTeam'].value_counts()
    return l

    
```

```python
a='Punjab Kings'
b='Mumbai Indians'
compare(a,b)
```

```python
### Created a function to know which team has won finals 
def winner(team):
    
    a=ipl[(ipl['MatchNumber'] == 'Final') & (ipl['WinningTeam']==team)]
    return a
p='Chennai Super Kings'
winner(p)
```
